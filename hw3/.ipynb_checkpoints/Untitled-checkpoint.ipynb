```python
import numpy as np
from scipy.linalg import lu

def PageRank(A,method):
    
    out_degrees = A.sum(axis=1)
    r = out_degrees
    p = 0.85
    n = A.shape[0]
    r_plus = np.where(r > 0, 0.85 / r, 0)
    z = np.where(r > 0, (1-p)/n, 1/n)
    z = z[:,np.newaxis]
    ones =  np.ones(n)[:,np.newaxis].T

    P = np.diag(r_plus) @ A.values  +  (z @ ones)
    
    inv_matrix = (np.eye(n) - P.T)
    inv_matrix[0] = 1

    zeros = np.zeros((n,n))
    zeros[0,0] = 1
    
    # LU decomposition
    if method == "LU":
        # Perform LU decomposition
        _, L, U = lu(inv_matrix)

        # Solve the linear equation using forward and backward substitutions
        y = np.linalg.solve(L, zeros)
        x = np.linalg.solve(U, y)
        
    # QR factorization
    if method == "QR":
        # Perform QR decomposition
        Q, R = np.linalg.qr(inv_matrix)

        # Solve the linear equation using back substitution
        x = np.linalg.solve(R, np.dot(Q.T, zeros))
    
    #Jacobi
    if method == "Jacobi":
        D = np.diag(inv_matrix)
        L = np.tril(inv_matrix)
        U = np.triu(inv_matrix) 
        b = np.zeros((n,n))
        
        D_inv = np.linalg.inv(D)
        T = D_inv @ (L + U)
        
        max_iterations = 1000
        x = x = np.zeros_like(b)
        
        for i in range(max_iterations):
            x_new = T @ x
            if np.allclose(x, x_new, atol=1e-10, rtol=0.):
                break
            x = x_new
        x = x_new
        
        # Set the maximum number of iterations and tolerance
        #max_iterations = 1000
        #tolerance = 1e-6

        # Initialize the solution vector x
        #zeros = np.zeros((n,n))
        #x = np.random.rand(n,n)

        # Perform the Jacobi method
        #for iteration in range(max_iterations):
        #    x_new = np.zeros_like(x)
        #    for i in range(inv_matrix.shape[0]):
        #        s = np.dot(inv_matrix[i, :i], x[:i]) + np.dot(inv_matrix[i, i+1:], x[i+1:])
        #        x_new[i] = (zeros[i] - s) / inv_matrix[i, i]
        #    if np.allclose(x, x_new, atol=tolerance):
        #        break
        #    x = x_new
    
    #SVD
    if method == "SVD":
        # Perform Singular Value Decomposition (SVD)
        U, s, Vh = np.linalg.svd(inv_matrix)

        # Solve the linear equation using back substitution
        y = np.linalg.solve(np.diag(s), np.dot(U.T, zeros))
        x = np.dot(Vh.T, y)
    
    return x


```
